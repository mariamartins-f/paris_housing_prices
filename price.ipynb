```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("syedaeman2212/paris-housing-prices")

print("Path to dataset files:", path)
```

```python
import pandas as pd
df = pd.read_csv("paris_housing_prices_dataset.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
df.plot(kind='scatter', x="Price_EUR", y="Size_sqm")
plt.show()
```

```python
correlacao=df["Price_EUR"].corr(df["Size_sqm"])
coef_determinacao=(correlacao**2)*100
print(f"Coeficiente de determinação: {coef_determinacao:.2f}%")
```

```python
import pandas as pd
from sklearn import linear_model

df = pd.read_csv("paris_housing_prices_dataset.csv")

modelo = linear_model.LinearRegression()
x = df[["Size_sqm"]]
y = df[["Price_EUR"]]

modelo.fit(x, y)

```

```python
print(modelo.predict([[89]])[0][0])
```
